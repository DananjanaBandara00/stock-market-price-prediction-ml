```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# === Paths ===
input_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\results\outputs\IT24101462.xlsx"
output_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\results\outputs\IT24101452.xlsx"

# === Load Dataset ===
df = pd.read_excel(input_path, index_col="Date")

# Example: normalize values
cols = ['Open','High','Low','Close*','Adj Close**','Volume']
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[cols])

for i, col in enumerate(cols):
    df[col + "_scaled"] = scaled[:, i]

print(df.head())
print("Step 4 done.")

print(df.head())
# === Save Output ===
df.to_excel(output_path)
print(f" Step Done → {output_path}")
```
