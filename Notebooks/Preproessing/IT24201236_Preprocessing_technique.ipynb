```python
import pandas as pd
import matplotlib.pyplot as plt

# === Paths ===
input_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\results\outputs\IT24101491.xlsx"
output_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\results\outputs\IT24201236.xlsx"

# === Load Dataset ===
df = pd.read_excel(input_path, index_col="Date")

df['ret_1d'] = df['Adj Close**'].pct_change().fillna(0)

corr = df.corr()
plt.imshow(corr.values, cmap="coolwarm", aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()

# === Save Output ===
df.to_excel(output_path)
print(f"Step Done → {output_path}")
```
