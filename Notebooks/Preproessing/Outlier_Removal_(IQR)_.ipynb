```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/IT24101341.xlsx',index_col="Date")
```

```python
#Create new feature
df["Log_Close"] = np.log(df["Adj Close**"])
```

```python
# === Outlier Removal for Volume ===
# Calculate Q1, Q3, and IQR
Q1 = df['Volume'].quantile(0.25)
Q3 = df['Volume'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter out outliers
df = df[(df['Volume'] >= lower_bound) & (df['Volume'] <= upper_bound)]
```

```python
# === Plot Boxplot of Volume ===
plt.boxplot(df['Volume'])
plt.title("Boxplot of Volume After Outlier Removal")
plt.show()
```

```python
# Save preprocessed dataset as Excel
df.to_excel("IT24101462.xlsx")
print("Preprocessed dataset saved as 'IT24101462.xlsx'")
```
