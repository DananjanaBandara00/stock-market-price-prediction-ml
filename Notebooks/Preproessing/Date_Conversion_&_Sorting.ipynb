```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/yahoo_data.xlsx')
```

```python
#Preprocessing
#Convert 'Date' to datetime for proper time-series handling
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Sort by date and set as index
df = df.sort_values('Date').set_index('Date')
```

```python
# Handle missing values
df = df.ffill().bfill()
```

```python
#Exploratory Data Analysis (EDA)
df['Adj Close**'].plot(title="Adjusted Close Over Time")
plt.xlabel("Date"); plt.ylabel("Adj Close")
plt.show()
```

```python
# Save preprocessed dataset as Excel
df.to_excel("IT24101357.xlsx")
print("Preprocessed dataset saved as 'IT24101357.xlsx'")
```
