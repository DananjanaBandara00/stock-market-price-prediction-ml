```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/IT24101452.xlsx', index_col="Date")
```

```python
# 1.Daily Returns
df['ret_1d'] = df['Adj Close**'].pct_change()
```

```python
# 2. Moving Averages
df['ma_7'] = df['Adj Close**'].rolling(window=7).mean()
df['ma_30'] = df['Adj Close**'].rolling(window=30).mean()
```

```python
# 3. 14-day Volatility
df['vol_14'] = df['ret_1d'].rolling(window=14).std()
```

```python
# Remove rows with NaN values caused by rolling calculations
df = df.dropna()
```

```python
# Plot Adjusted Close with Moving Averages
df[['Adj Close**', 'ma_7', 'ma_30']].plot(title="Adjusted Close with Moving Averages")

plt.show()
```

```python
#Reset index so Date becomes a column again
df = df.reset_index()
```

```python
# Save preprocessed dataset as Excel
df.to_excel("IT24101491.xlsx", index=False)
print("Preprocessed dataset saved as 'IT24101491.xlsx'")
```
