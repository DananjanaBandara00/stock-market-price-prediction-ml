```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# === Paths ===
input_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\data\yahoo_data.xlsx"
output_path = r"C:\Users\pulina\Desktop\2025-Y2-S1-MLB-B6G2-01\results\outputs\IT24101357.xlsx"

df = pd.read_excel(input_path)

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').set_index('Date')
df = df.ffill().bfill()   # Handle missing values

# === Save Output ===
df.to_excel(output_path)
print(f"Step Done → {output_path}")
# Visualization
df['Adj Close**'].plot(title="Adjusted Close Over Time")
plt.xlabel("Date"); plt.ylabel("Adj Close")
plt.show()
```
